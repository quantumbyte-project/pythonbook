Como una herramienta visual, Matplotlib es una de las librerías de Python orientada a la representación dinámica y estática a partir de datos proporcionados.

Matplotlib cuenta con una extensa cantidad de módulos, sin embargo, nos enfocaremos en el el módulo ***pyplot*** el cual está diseñado para casos sencillos de gráficos en programación y para que éstos sean interactivos. Éste módulo nos permite generar gráficas en dos ejes a partir de la información que brinda el usuario por medio de _listas (o arrays)_, por lo que también se emplea la extensión matemática de Numpy.



***
## Sintáxis para pyplot

Con la finalidad de mostrar trazos en la pantalla, Matplotlib se basa en crear objetos llamados figuras, que son ilustraciones completas y cuyas propiedades se pueden modificar con el objetivo de que el usurio personalice la graficación a su gusto.

En esta librería, una figura es una ilustración completa y cada uno de los trazados de dicha figura se denomina eje. Crear una gráfica es el acto de hacer plotting

Como primer recurso, se debe importar la librería Numpy para el manejo de listas y el módulo _**pyplot**_ de matplotlib para la creación de trazos, éste último con regularidad se encontrará renombrado como _**plt**_:

```python
import numpy as np
import matplotlib.pyplot as plt
```


La parte fundamental de este código es crear una figura que tenga dos ejes:
```python
fig,ax=plt.subplots() 
```
_**Nota:** ¡Preste atención que en el código anterior, "**fig**" va seguido de una **coma( *`,`* )**!_

Donde:
* *`fig,ax`*⟶Es el operador que crea una figura (**fig**) con ejes (**ax**, por la abreviación en inglés de "_axes_").
* *`plt.subplots()`*⟶Es el comando de pyplot para generar subgráficas.

Y por último, la línea de código que se encargará de graficar será la siguiente:

```python
ax.plot([lista en x],[lista en y])
```
Donde:
* *`ax.plot()`*⟶Es la función que se encargará de realizar los trazos dentro de los ejes estándares.
* *`([lista en x],[lista en y])`*⟶Son las coordenadas en *`x`* y en *`y`*, cuyos resultados se obtienen de listas.

El uso de las listas es equivalente a la tabulación para una gráfica manual; donde las coordenadas quedan definidas por la correspondencia directa entre *`x`* y *`y`*. Es decir, supongamos la siguiente relación:

| x | y |
|:--|:--|
| -3| 7 |
| -2| 5 |
| -1| 3 |
| 0 | 0 |
| 1 | 2 |
| 2 | 6 |
| 3 | 8 |

Cuando se grafique, los puntos que se tomarán serán *`(-3,7),(-2,5),(-1,3),(0,0),(1,2),(2,6),(3,8)`*. Por otro lado, gracias a lo visto en temas pasados, sabemos que, tanto la columna de *`x`* como *`y`* pueden expresarse en forma de listas, siendo estas *`x=`[-3,-2,-1,0,1,2,3]`* y *`y=[7,5,3,0,2,6,8]`*. Por lo que, al retomar la instrucción **`ax.plot([lista en x],[lista en y])`**, obtendremos una relación de los puntos **`([-3,-2,-1,0,1,2,3],[7,5,3,0,2,6,8])`** tiene ahora una correspondencia respectiva.

De esta forma, nuestro código completo quedaría como:

````python
import numpy as np
import matplotlib.pyplot as plt

fig,ax=plt.subplots()
ax.plot([-3,-2,-1,0,1,2,3],[7,5,3,0,2,6,8])
````

Lo cual, nos permite concluir nuestro primer ejemplo.

## Configuración de gráficos

Hasta el momento, conocemos la sintáxis básica que nos permite nada más que graficar. Sin embargo, es posible modificar las propiedades de esta para brindar más información o ser más específicos con ellos; pyplot cuenta con más comandos que nos permiten esto, los cuales, no importa el orden en el que se coloquen, solo asegurándose la intrucción sea declarada posterior a la creación de la figura:

* *`ax.grid()`*: Esta función crea una cuadrícula o rejilla para la figura.

````python
import numpy as np
import matplotlib.pyplot as plt

fig,ax=plt.subplots()
ax.grid()
ax.plot([-3,-2,-1,0,1,2,3],[7,5,3,0,2,6,8])
````

### Títulos y texto

Estos comandos toman como argumento cadenas de caracteres con el principal propósito de imprimirlos en la figura:

* *`ax.set_tittle("Título de la gráfica")`*: Con este comando, se le puede dar un título a la gráfica.

* *`ax.set_xlable("Nombre eje x")`* y *`ax.set_ylable("Nombre eje y")`*: Ambas cumplen la misma función de darle un título a sus respectivos ejes a partir de una cadena de caracteres dada.

* *`ax.plot(argumento_1,argumento_2,label="Nombre del trazo")`* y *`ax.legend()`*: El primero es una extención de la función *`ax.plot()`*, cuyo tercer argumento se encarga de otorgarle un nombre a la función graficada, sin embargo, ésta no se imprimirá sin la función ***`ax.legend()`***.

* *`ax.text( Posición en x , posición en y , "Texto" , fontsize = n )`*: Por este medio, será posible escribir cadenas de caracteres en cualquier punto especificado en la figura; la función recibe 4 parámetros, donde ***`Posición en x, posición en y`*** hace referencia a la coordenada donde se pondrá la cadena de caracteres, ***`"Texto"`*** es la cadena que se imprimirá, y ***`fontsize=n`*** es el tamaño de fuente que se desea tener.

**Nota:** Si en cualquiera de los casos se desea escribir símbolos algebráicos o terminología de LaTeX, en general, debe escribirse de esta manera ***`r"Texto $ _LaTeX_"$`***.

````python
import numpy as np
import matplotlib.pyplot as plt

fig,ax=plt.subplots()
ax.grid()
ax.set_title("Título de la gráfica")
ax.set_xlabel("Nombre del eje x")
ax.set_ylabel("Nombre del eje y")
ax.plot([-3,-2,-1,0,1,2,3],[7,5,3,0,2,6,8],label="Nombre del trazo")
ax.legend() #¡Importante para imprimir el título anterior!
ax.text(0,5,"Texto",fontsize=15)
ax.text(0,0,r"$\int_{a}^{b}$",fontsize=20) #Para escribir con LaTeX
````

### Color

Si se desea cambiar el color de algo impreso en la figura, se puede realizar solo extendiendo el argumento de la función a modificar, es decir:

* *`funcion(arg_1,...,color="Nombre del color")`*: La etensión del argumento se puede realizar en cualquiera de los comandos vistos —a excepción de *`ax.legend()`*— háblese de la rejilla, las gráficas, el texto, los títulos, etc. Un dato importante es que los colores que acepte este argumento deben estar escritos en ***inglés***

````python
import numpy as np
import matplotlib.pyplot as plt

fig,ax=plt.subplots()
ax.grid(color="darkblue")
ax.set_title("Título de la gráfica",color="red")
ax.set_xlabel("Nombre del eje x",color="red")
ax.set_ylabel("Nombre del eje y",color="red")
ax.plot([-3,-2,-1,0,1,2,3],[7,5,3,0,2,6,8],label="Nombre del trazo",color="aqua")
ax.legend() #¡Importante para imprimir el título anterior!
ax.text(0,5,"Texto",fontsize=15,color="purple")
ax.text(0,0,r"$\int_{a}^{b}$",fontsize=20,color="purple") #Para escribir con LaTeX
````

### Tipos de gráficos

A pesar de que hasta el momento solo se ha trabajado con un tipo de trazado, la librería cuenta con otras más opciones.

* *`plot(x,y)`*: La función que hasta el momento se ha utilizado recibe como parámetros una lista en *`x`* y en *`y`*.

* *`scatter(x,y)`*: Con este comando es posible utilizar la gráfica de dispersión, el cual necesita como argumento, igualmente, una lista en *`x`* y en *`y`*.

````python
import numpy as np
import matplotlib.pyplot as plt

fig,ax=plt.subplots()
ax.grid()
ax.set_title(r"Ejemplo $scatter(x,y)$ ")
ax.scatter([-3,-2,-1,0,1,2,3],[7,5,3,0,2,6,8])
````

* *`bar(x,height)`*: Podemos utilzar la gráfica de barras con este comando, cuyos parámetros es una lista de *`x`* y una algura para casa una que puede tomarse, igualmente, como una lista en *`y`*.

````python
import numpy as np
import matplotlib.pyplot as plt

fig,ax=plt.subplots()
ax.grid()
ax.set_title(r"Ejemplo $bar(x,height)$ ")
ax.bar([-3,-2,-1,0,1,2,3],[7,5,3,0,2,6,8])
````

* *`pie(x)`*: Ésta es la característica gráfica de pastel, la cual, a diferencia de las demás, solo necesita una lista en *`x`*, el mismo programa se encarga de distribuir la gráfica a partir de los datos. No está de más recordar que al necesitar números contables, debemos evitar el uso de números negativos.

````python
import numpy as np
import matplotlib.pyplot as plt

fig,ax=plt.subplots()
ax.grid()
ax.set_title(r"Ejemplo $bar(x,height)$ ")
ax.pie([0,10,24,33])
````

### Subgráficas

Las subgráficas en python son gráficas indepentientes creadas en una misma figura. Su manejo se basa solamente en crear más de ***un par de ejes estándares***, es decir, al momento de crear la figura tendremos:

```python
fig,(ax1,ax2,...,axn)=plt.subplots(filas,columnas)
```
Donde:
* *´(ax1,ax2,...,axn)´*: es la declaración de la cantidad de gráficas —que existen en sus respectivos ejes— que se mostrarán en la pantalla.
* *´plt.subplots(filas,columnas)´*: es la función de pyplot —*que hasta el momento no se ha usado*— que nos permite realizar subgráficas y cuyos argumentos son la cantidad de filas y columnas en las que estarán organizados.

El resto de los comandos se mantiene igual, solo es necesario prestar atención a cuál de los ejes estamos llamando (ax1,ax2,...,axn).

````python
import numpy as np 
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi, 2*np.pi, 1001)

f = np.sin(x)
g = np.cos(x)
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1,4, figsize=(16,4))

#Configurando las propiedades para la primer gráfica
ax1.set_title("Gráfica 1")
ax1.plot(x, f)
ax1.grid()

#Configurando las propiedades para la segunda gráfica
ax2.plot(x, g)
ax2.grid()

#Configurando las propiedades para la tercera gráfica
ax3.plot(x, x**2)
ax3.grid()

#Configurando las propiedades para la cuarta gráfica
ax4.plot(x, x**3 -5, color="orange")
ax4.grid()
````

## Ejemplos

### Graficar más de una función

Supongamos tres diferentes funciones que se desean graficar:
$$f(x)=x$$
$$g(x)=x^2$$
$$h(x)=x^3$$

Por conocimientos previos, se sabe que el valor que retorne *`f(x)`* *—tomando el primer ejempo—* corresponderá al valor de *`y`*; por tanto, nuestra coordenada propuesta estará dada por *`(x,f(x))`*. Si bien, tenemos el comportamiento que tendrá cada valor de *`y`*, hasta el momento no se ha definido el comportamiento que tendrá *`x`*, es decir, no sabemos cuántos valores evauaremos ni en qué límites estarán. Para que el trazo de la función tenga la continuidad que deseamos, *`x`* deberá tomar muchos valores consecutivos, dentro de un intervalo establecido (sea grande o pequeño).

De este modo, definiremos un vector que esté en el intervalo de *`[0,2]`* y que contenga *`1000`* valores.

```python
x = np.linspace(0,2,1001)
```

Después, crearemos la figura con los ejes estándares (*`fig,ax`*), y de manera prosiguiente trazaremos las funciones. En este punto, es importante recalcar que si escribimos más de una vez el comando *`ax.plot()`*, el programa lo interpretará como imprimir varias figuras dentro de una misma gráfica.

```python
ax.plot(x,x)
ax.plot(x,x**2)
ax.plot(x,x**3)
```

Por tanto, el código funcional quedará como:

````python
x = np.linspace(0,2,1001)
fig,ax=plt.subplots()
ax.plot(x,x,label="f(x)") #f(x)
ax.plot(x,x**2,label="g(x)") #g(x)
ax.plot(x,x**3,label="h(x)") #h(x)
ax.legend()
ax.grid()
ax.set_title("Gráfica de varias funciones")
ax.set_xlabel("Eje x")
ax.set_ylabel("Eje y")
````

### Funciones trigonométricas

Supongamos en este caso dos funciones trigonométricas:
$$f(x)=Sen(x)$$
$$g(x)=Cos(x)$$

Donde el dominio lo definiremos como:

```python
x=np.linspace(-2*np.pi,2*np.pi,1001)
```
Y en este caso, renombraremos las funciones bajo una variable:


```python
f=np.sin(x)
g=np.cos(x)
```

De esta manera cuando demos la instrucción de graficar no será necesario volver a escribir la función, y en este caso, le daremos color a las gráficas, al igual que un título:

```python
ax.plot(x, f, color="Red", label=r"$\sin(x)$")
ax.plot(x, g, color="purple", label=r"$\cos(x)$")
```

Por tanto, el código funcional quedará como:

````python
x = np.linspace(-2*np.pi, 2*np.pi, 1001)

f = np.sin(x)
g = np.cos(x)

fig, ax = plt.subplots()

ax.grid()
ax.set_title(r"Funciones trigonométricas")
ax.plot(x, f, color="Red", label=r"$\sin(x)$")
ax.plot(x, g, color="purple", label=r"$\cos(x)$")
ax.set_xlabel("Eje x")
ax.set_ylabel("Eje y")
ax.legend()
````
